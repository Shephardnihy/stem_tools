```python
import xarray
import numpy as np
import stem_tools
import matplotlib.pyplot as plt
import libertem.api as lt
from skimage.transform import downscale_local_mean
```

```python
from scipy.io import loadmat
```

```python
%matplotlib widget
```

# Load Data

```python
ctx = lt.Context()
```

```python
ds = ctx.load("frms6", path = "FGT512_Lorentz_049.hdr")
```

```python
gainmap = loadmat("GainMap_AB_B4_G001.mat")["GainMap"]
```

# Preprocess

```python
preprocess = stem_tools.UDFs.PreprocessFRMS6(gainmap = gainmap, 
                                             hardware_bin = 4,)
```

```python
preprocessed = ctx.run_udf(dataset = ds,
                           udf = preprocess,
                           progress = True)
```

```python
preprocessed = preprocessed["pattern"].data
```

```python
preprocessed.shape
```

# Save Data as h5 file

```python
metadata = stem_tools.io.get_metadata_from_hdr("FGT512_Lorentz_049.hdr")
```

```python
datafile = stem_tools.io.numpy_to_xarray(preprocessed, ("y", "x", "qy", "qx"), attrs = metadata)
```

```python
datafile.to_netcdf(path = "data.h5")
```
